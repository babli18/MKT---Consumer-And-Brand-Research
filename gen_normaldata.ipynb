```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# 1. Normal Distribution
mean = 0
std_dev = 1
normal_data = np.random.normal(mean, std_dev, 1000)

# 2. Binomial Distribution
n = 10  # number of trials
p = 0.5  # probability of success
binomial_data = np.random.binomial(n, p, 1000)

# 3. Poisson Distribution
lambda_poisson = 3  # rate parameter (average number of events)
poisson_data = np.random.poisson(lambda_poisson, 1000)

# 4. Exponential Distribution
lambda_exp = 1  # rate parameter (time between events)
exponential_data = np.random.exponential(1 / lambda_exp, 1000)

# 5. Chi-Square Distribution
df_chi = 4  # degrees of freedom
chi_square_data = np.random.chisquare(df_chi, 1000)

# 6. T-Distribution
df_t = 10  # degrees of freedom
t_distribution_data = np.random.standard_t(df_t, 1000)

# Plot the distributions
fig, axs = plt.subplots(3, 2, figsize=(12, 15))

# Normal Distribution Plot
sns.histplot(normal_data, kde=True, ax=axs[0, 0], color='blue')
axs[0, 0].set_title('Normal Distribution')

# Binomial Distribution Plot
sns.histplot(binomial_data, kde=False, ax=axs[0, 1], color='green')
axs[0, 1].set_title('Binomial Distribution')

# Poisson Distribution Plot
sns.histplot(poisson_data, kde=False, ax=axs[1, 0], color='purple')
axs[1, 0].set_title('Poisson Distribution')

# Exponential Distribution Plot
sns.histplot(exponential_data, kde=True, ax=axs[1, 1], color='red')
axs[1, 1].set_title('Exponential Distribution')

# Chi-Square Distribution Plot
sns.histplot(chi_square_data, kde=True, ax=axs[2, 0], color='orange')
axs[2, 0].set_title('Chi-Square Distribution')

# T-Distribution Plot
sns.histplot(t_distribution_data, kde=True, ax=axs[2, 1], color='cyan')
axs[2, 1].set_title('T-Distribution')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

# Sample data: lifetimes of 10 randomly selected light bulbs
sample_data = [1020, 980, 1005, 995, 1010, 1003, 999, 1007, 1015, 990]

# Population mean (standard lifetime)
pop_mean = 1000

# Calculate the t-test for one sample
t_statistic, p_value = stats.ttest_1samp(sample_data, pop_mean)

# Calculate degrees of freedom
n = len(sample_data)
df = n - 1

# Print results
print(f"Sample Mean: {np.mean(sample_data)}")
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {df}")

# Check if the result is significant at a 5% significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The sample mean is significantly different from 1000 hours.")
else:
    print("Fail to reject the null hypothesis. The sample mean is not significantly different from 1000 hours.")
```
